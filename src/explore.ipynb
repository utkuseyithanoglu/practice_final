```python
import requests
import pandas as pd


url = "https://data.cityofnewyork.us/resource/76xm-jjuj.json"
params = {
    "$limit": 6000,
    "$where": "incident_datetime >= '2024-01-01'",
    "$order": "incident_datetime DESC"}
response = requests.get(url, params=params)
data = response.json()

df_2 = pd.DataFrame(data)
```
